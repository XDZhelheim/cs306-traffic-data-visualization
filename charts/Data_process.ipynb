```python
f = open("d_time_dist.csv","r")  

hr_cnt = {
    '0-6': [0 for col in range(5)],
    '6-12':[0 for col in range(5)],
    '12-18': [0 for col in range(5)],
    '18-24': [0 for col in range(5)],
    
}

hr_pr = {
    '0-6': [0 for col in range(5)],
    '6-12':[0 for col in range(5)],
    '12-18': [0 for col in range(5)],
    '18-24': [0 for col in range(5)],
    
}


hr_map = {'0':'0-6', '1':'0-6','2':'0-6','3':'0-6','4':'0-6','5':'0-6',\
          '6':'6-12','7':'6-12','8':'6-12','9':'6-12','10':'6-12','11':'6-12', \
          '12':'12-18','13':'12-18','14':'12-18','15':'12-18','16':'12-18','17':'12-18',\
          '18':'18-24','19':'18-24','20':'18-24','21':'18-24','22':'18-24','23':'18-24'}

while True:
    line = f.readline()
    if not line:
        break
    ele = line.split(',')
    hr_group = hr_map[str(int(ele[0]))]
    km = float(ele[1])
    
    
    if km <= 1:
        hr_cnt[hr_group][0] += 1
    elif 1 < km <= 5:
        hr_cnt[hr_group][1] += 1
    elif 5 < km <= 10:
        hr_cnt[hr_group][2] += 1
    elif 10 < km <= 30:
        hr_cnt[hr_group][3] += 1
    elif 30 < km:
        hr_cnt[hr_group][4] += 1


for key in hr_cnt:
    total1 = 0
    total2 = 100
    for num in hr_cnt[key]:
        total1 += num
    
    for i in range(4):
        hr_pr[key][i] = int(hr_cnt[key][i] / total1 * 100)
    
    for i in range(4):
        total2 -= hr_pr[key][i]
    hr_pr[key][4] = total2
```

```python
import numpy as np
import matplotlib.pyplot as plt
category_names = ['<1km', '1-5kms',
                  '5-10kms', '10-30kms', '>30kms']

hr_pr = {
    '0-6': [28, 48, 13, 8, 3],
    '6-12': [23, 53, 13, 8, 3],
    '12-18': [26, 51, 12, 8, 3],
    '18-24': [23, 54, 13, 8, 2],
    
}


def survey(results, category_names):
    """
    Parameters
    ----------
    results : dict
        A mapping from question labels to a list of answers per category.
        It is assumed all lists contain the same number of entries and that
        it matches the length of *category_names*.
    category_names : list of str
        The category labels.
    """
    labels = list(hr_pr.keys())
    data = np.array(list(hr_pr.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = plt.get_cmap('RdYlGn')(
        np.linspace(0.15, 0.85, data.shape[1]))

    fig, ax = plt.subplots(figsize=(9.2, 5))
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        rects = ax.barh(labels, widths, left=starts, height=0.5,
                        label=colname, color=color)

        r, g, b, _ = color
        text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        ax.bar_label(rects, label_type='center', color=text_color)
    ax.legend(ncol=len(category_names), bbox_to_anchor=(0, 1),
              loc='lower left', fontsize='small')

    return fig, ax


survey(hr_pr, category_names)
plt.show()
```

```python
import pandas as pd
import haversine as hs

data = pd.read_csv('OD_distance_data.csv')
def dist_func (location2,row):
    loc1 = (float(row['lng1']),float(row['lat1']))
    return float(hs.haversine(loc1,location2))

location = (114,22.5)
def FuncForDongWang(location):
    
    data['near_distance'] = data.apply (lambda row: dist_func(location,row), axis=1)
    data.drop(data[data.near_distance > 5].index, inplace=True)
    total = data['distance'].sum()
    row_count = data.shape[0]
    if row_count == 0:
        return "not enough data!"
    else:
        return total/row_count      # to string if needed
    
```
